```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(0, 2 * np.pi, 400)
y1 = np.sin(x ** 2)
y2 = np.cos(x ** 2)

fig, (ax1, ax2) = plt.subplots(1, 2)  # 1 row, 2 columns
fig.suptitle('Horizontally Stacked Subplots')

# First subplot
ax1.plot(x, y1, 'r-')  # Red solid line
ax1.set_title('Square of sine')

# Second subplot
ax2.scatter(x, y2, color='blue')  # Blue scatter plot
ax2.set_title('Square of cosine')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D  # Import the 3D toolkit

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot a 3D surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add a color bar which maps values to colors.
fig.colorbar(surface, shrink=0.5, aspect=5)

ax.set_title('3D Surface Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)                # sine values

plt.figure()  # Create a new figure
plt.plot(x, y)  # Plot the sine wave
plt.title('Sine Wave')  # Add a title
plt.xlabel('x')  # Label the x-axis
plt.ylabel('sin(x)')  # Label the y-axis
plt.grid(True)  # Show grid
plt.show()  # Display the plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

x = np.linspace(0, 10, 100)
y = np.sin(x)
line, = ax.plot(x, y, 'g-', picker=5)  # "picker=5" means a click within 5 points

def onpick(event):
    thisline = event.artist
    xdata = thisline.get_xdata()
    ydata = thisline.get_ydata()
    ind = event.ind
    points = tuple(zip(xdata[ind], ydata[ind]))
    print('on pick points:', points)

fig.canvas.mpl_connect('pick_event', onpick)

plt.show()
```

```python
import plotly.graph_objects as go

import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines+markers'))

# Update the layout
fig.update_layout(title='Interactive Plot of Sine Wave',
                  xaxis_title='x',
                  yaxis_title='sin(x)')

# Show the figure
fig.show()

```
